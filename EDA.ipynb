```python
#distribution de x_wins et x_draw
df = pd.read_csv("dataset.csv")
df.head()
df.info()
df.columns
df["x_wins"].value_counts()
df["is_draw"].value_counts()
print(df["x_wins"].value_counts(normalize=True) * 100)
print(df["is_draw"].value_counts(normalize=True) * 100)

fig, axes = plt.subplots(1,2, figsize=(10,4))

df["x_wins"].value_counts().plot(kind="bar", ax=axes[0], title="x_wins")
df["is_draw"].value_counts().plot(kind="bar", ax=axes[1], title="is_draw")

plt.show()


```

```python
#heatmap de corellation entre featurs et cibles
corr_xwins = df.corr(numeric_only=True)["x_wins"].sort_values(ascending=False)
print(corr_xwins)
corr_draw = df.corr(numeric_only=True)["is_draw"].sort_values(ascending=False)
print(corr_draw)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Heatmap corrélation features vs cibles")
plt.show()

targets_corr = df.corr(numeric_only=True)[["x_wins","is_draw"]]

plt.figure(figsize=(6,8))
sns.heatmap(targets_corr, annot=True, cmap="coolwarm")
plt.title("Corrélation features avec x_wins et is_draw")
plt.show()
```
